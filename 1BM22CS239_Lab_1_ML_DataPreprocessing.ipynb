```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Step 2: Load the Diabetes dataset
diabetes_df = pd.read_csv("/content/Dataset of Diabetes .csv")

# Step 3: Check for missing values
print("Missing values in Diabetes dataset:\n", diabetes_df.isnull().sum())

# Step 4: Handle missing values
# ----> Updated to only fill numeric columns with the mean
# Fill missing numerical values with mean for numeric columns only
numeric_cols = diabetes_df.select_dtypes(include=np.number).columns
diabetes_df[numeric_cols] = diabetes_df[numeric_cols].fillna(diabetes_df[numeric_cols].mean())

# Fill missing categorical values with mode
for col in diabetes_df.select_dtypes(include=['object']).columns:
    diabetes_df[col].fillna(diabetes_df[col].mode()[0], inplace=True)

# Step 5: Identify categorical columns and encode them
categorical_cols = ['Gender', 'CLASS']  # Adjust based on actual column names
label_enc = LabelEncoder()
for col in categorical_cols:
    diabetes_df[col] = label_enc.fit_transform(diabetes_df[col])

# Step 6: Apply Min-Max Scaling
min_max_scaler = MinMaxScaler()
diabetes_df_scaled = pd.DataFrame(min_max_scaler.fit_transform(diabetes_df), columns=diabetes_df.columns)

# Step 7: Apply Standardization
std_scaler = StandardScaler()
diabetes_df_standardized = pd.DataFrame(std_scaler.fit_transform(diabetes_df), columns=diabetes_df.columns)

# Step 8: Save the preprocessed datasets
diabetes_df_scaled.to_csv("/content/diabetes_preprocessed.csv", index=False)

# Create the 'content' directory if it doesn't exist
os.makedirs("content", exist_ok=True)

diabetes_df_standardized.to_csv("content/diabetes_standardized.csv", index=False)

print("Diabetes dataset preprocessing completed. Preprocessed datasets saved.")
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Step 2: Load the Adult Income dataset
adult_df = pd.read_csv("/content/adult.csv")

# Step 3: Check for missing values
print("Missing values in Adult Income dataset:\n", adult_df.isnull().sum())

# Step 4: Handle missing values
# Fill missing categorical values with mode
for col in adult_df.columns:
    if adult_df[col].isnull().sum() > 0:
        adult_df[col].fillna(adult_df[col].mode()[0], inplace=True)

# Step 5: Identify categorical columns and encode them
categorical_cols = adult_df.select_dtypes(include=['object']).columns
label_enc = LabelEncoder()
for col in categorical_cols:
    adult_df[col] = label_enc.fit_transform(adult_df[col])

# Step 6: Apply Min-Max Scaling
min_max_scaler = MinMaxScaler()
adult_df_scaled = pd.DataFrame(min_max_scaler.fit_transform(adult_df), columns=adult_df.columns)

# Step 7: Apply Standardization
std_scaler = StandardScaler()
adult_df_standardized = pd.DataFrame(std_scaler.fit_transform(adult_df), columns=adult_df.columns)

# Step 8: Save the preprocessed datasets
adult_df_scaled.to_csv("/content/adult_preprocessed.csv", index=False)

# Create the 'content' directory if it doesn't exist
os.makedirs("content", exist_ok=True)

adult_df_standardized.to_csv("content/adult_standardized.csv", index=False)

print("Adult Income dataset preprocessing completed. Preprocessed datasets saved.")
```
